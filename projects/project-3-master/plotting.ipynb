```python
df_scores = pd.read_csv('results/scores_all.csv', index_col=0)
```

```python
df_scores_lemma = pd.read_csv('results/scores_lemma.csv', index_col=0)
```

```python
df_all = df_scores.join(df_scores_lemma)
```

```python
df_all.sort_values(by='Test Score', ascending=False)
```

```python
df_all[['Test Score', 'Test Score-lemma']].T
```

```python
plt.figure(figsize=(15,8))
plt.title('Classification: Test Score per Model\nBlue: Normal, Tan: Lemma', fontsize=20)
sns.barplot(data=df_all[['Test Score']].T, alpha=0.5, color='blue')
sns.barplot(data=df_all[['Test Score-lemma']].T, alpha=0.5, color='orange')
plt.xticks(fontsize=15, rotation=-20, horizontalalignment='left');
```
